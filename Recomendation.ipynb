```python
import ast
#ast.literal_eval()

def convert(obj):
    L = []
    
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return(L)
        
convert(movies.iloc[0].genres)    
```

```python
import ast
#ast.literal_eval()

def convert3(obj):
    L = []
    counter = 0
    
    
    for i in ast.literal_eval(obj):
        if counter != 3:
            counter+=1
            L.append(i['name'])
    return(L)

movies['cast'] = movies['cast'].apply(convert3)
#convert3(movies.iloc[0].cast)
```

```python
def convertCrew(obj):
    L = list()
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return i['name']
        
convertCrew(movies.iloc[3].crew)
movies['crew'] = movies['crew'].apply(convertCrew)          
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer as CV
cv = CV(max_features = 5000,stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names() #in-build method to get all feature names. Just to check.
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cos
```

```python
similarity=cos(vectors)
similarity[0]
sorted(list(enumerate(similarity[0])),reverse  = True) 


#similarity contains the similar score of movies with each other. Now 
```

```python
sorted(list(enumerate(similarity[1])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
new_df[new_df['title'] == 'John Carter'].index[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances  =  similarity[movie_index]
    movies_list  = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```
